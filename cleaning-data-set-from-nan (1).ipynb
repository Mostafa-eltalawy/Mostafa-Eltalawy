* counting number of NAN in each column

```python
Total=df.isnull().sum().sort_values(ascending=False)
```

* calculating percentage of NAN in each column

```python
percentage=(df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
```

* make concat for percentage and total numbers of NAN

```python
null_cal=pd.concat([Total,percentage],axis=1,keys=["nul_Total","nul_percentage"])
```

```python
null_cal
```

* deroping all columns that have NAN percentage is greater than 80%

```python
new_df = df.drop((null_cal[null_cal['nul_percentage'] > 0.2]).index,1)
```

```python
new_df.shape
```

```python
new_df.head()
```

* checking nulls for new data frame

```python
total_new=new_df.isnull().sum().sort_values(ascending=False)
```

```python
total_new
```

* replacing all NAN with 0 value

```python
new_df=new_df.fillna(0)
```

* checking NAN for final data frame after replacing all NAN with 0 value**

```python
NAN_final_check=new_df.isnull().sum().sort_values(ascending=False)
```
